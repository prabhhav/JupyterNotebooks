```
import numpy as np
import pandas as pd
import numpy.matlib
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```
def calc_distance(X1, X2):
    return (sum((X1 - X2)**2))**0.5
    
```

```
def calc_centroids(clusters, X):
    new_centroids = []
    new_df = pd.concat([pd.DataFrame(X), pd.DataFrame(clusters, columns=['cluster'])],
                      axis=1)
    for c in set(new_df['cluster']):
        current_cluster = new_df[new_df['cluster'] == c][new_df.columns[:-1]]
        cluster_mean = current_cluster.mean(axis=0)
        new_centroids.append(cluster_mean)
    return new_centroids
```

```
def findClosestCentroids(ic, X):
    assigned_centroid = []
    for i in X:
        distance=[]
        for j in ic:
            distance.append(calc_distance(i, j))
        assigned_centroid.append(np.argmin(distance))
    return assigned_centroid
```

Mini Code

```
df_mini = df1.sample(30)
df_mini.reset_index(inplace=True)
```

```
df_mini.drop('index',axis=1,inplace=True)
df_mini
```

```
plt.figure()
plt.scatter(df_mini[0], df_mini[1])
plt.show()
```

Step 1
* Choose value of K, here k=3
* Choose k(3) random points as cluster heads

```
init_centroids = random.sample(range(0, len(df_mini)),3)
init_centroids
```

```
centroids = []
for i in init_centroids:
    centroids.append(df_mini.loc[i])
centroids = np.array(centroids)
centroids
```

Step 2
* Calculate the Euclidean distance and group datapoints to the nearest centroid

```
print("{:<20} {:<20} {:<20} {:<20} {:<20} {:<20}".format("x1", "x2", "C1 Dist", "C2 Dist", "C3 Dist", "C Head"))

cluster = []

for index, row in df_mini.iterrows():
    distance = []
    for i, c in enumerate(centroids):
      distance.append(calc_distance(c, row.to_numpy()))
    print("{:<20} {:<20} {:<20} {:<20} {:<20} {:<20}".format(row[0], row[1], distance[0], distance[1], distance[2], np.argmin(distance)))
    cluster.append(np.argmin(distance))
```

```
plt.figure()
plt.scatter(df_mini[0], df_mini[1], c=cluster)
plt.show()
```

```
cluster0_groups = df_mini[np.array(cluster) == 0]
```

```
cluster0_groups
```

```
cluster1_groups = df_mini[np.array(cluster) == 1]
```

```
cluster2_groups = df_mini[np.array(cluster) == 2]
```

```
new_centroids = [np.array(cluster0_groups.mean()), np.array(cluster1_groups.mean()), np.array(cluster2_groups.mean())]
```
